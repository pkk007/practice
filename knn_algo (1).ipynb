```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('/content/drive/My Drive/knn/KNN_Data_Set_v2.0.csv')
print("shape of data set : ",df.shape)
print("columnbs : ",df.columns)
df.describe()
```

```python
plt.xlabel('aptitude')
plt.ylabel('communication')
plt.scatter(df.Aptitude, df.Communication, color = 'blue')
plt.show()
```

```python
#now train and test the data set
from sklearn.model_selection import train_test_split

predictors = df.iloc[:,1:3]
target =  df.iloc[:,3]
training_data, testing_data = train_test_split(df, test_size = 0.3, random_state = 0)
train_X, test_X, train_y, test_y = train_test_split(predictors, target , test_size=0.1, random_state =0)
print(train_X)
print(test_X)
print(train_y)
print(test_y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN_Classifier = KNeighborsClassifier(n_neighbors=3)
KNN_Classifier.fit(train_X,train_y)
print("Test set predictions : {}".format(KNN_Classifier.predict(test_X)))
```

```python
from sklearn.metrics import accuracy_score
accuracy =  KNN_Classifier.score(test_X,test_y)
print(accuracy)

from sklearn.metrics import confusion_matrix
predictors = KNN_Classifier.predict(test_X)
cnf_matrix = confusion_matrix(test_y, predictors)
print(cnf_matrix)
```
