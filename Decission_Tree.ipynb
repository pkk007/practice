```python
import numpy as np
import pandas as pd

df = pd.read_csv('/content/drive/My Drive/MACHINE LEARNING/DT.csv')
df.shape
df.columns
df
```

```python
from sklearn.preprocessing import LabelEncoder
lblEncod = LabelEncoder()
encode_df = df.apply(lblEncod.fit_transform)
print(encode_df)
```

```python
from sklearn.model_selection import train_test_split
predictors = encode_df.iloc[:,0:4]
target = encode_df.iloc[:,4]
train_X, test_X, train_y, test_y = train_test_split(predictors, target, test_size=0.1, random_state=100)
print(test_X)
print(test_y)
```

```python
from sklearn.tree import DecisionTreeClassifier
DT_classifier = DecisionTreeClassifier(criterion='entropy', random_state=100, max_depth=4, min_samples_leaf=1)
DT_classifier.fit(train_X, train_y)
print("Test set Predictions : {}".format(DT_classifier.predict(test_X)))
```

```python
from sklearn.metrics import accuracy_score
accuracy = DT_classifier.score(test_X, test_y)
print(accuracy)
from sklearn.metrics import confusion_matrix
predictions = DT_classifier.predict(test_X)
cnf_matrix = confusion_matrix(test_y, predictions)
print(cnf_matrix)
```
